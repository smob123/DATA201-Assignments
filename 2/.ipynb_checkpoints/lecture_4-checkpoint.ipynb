```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
oecd_bli = pd.read_csv("oecd_bli_2015.csv", thousands=',')
```

```python
oecd_bli.head()
```

```python
oecd_bli.info()
```

```python
oecd_bli[oecd_bli["Country"]=="New Zealand"]
```

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
```

```python
oecd_bli.head()
```

```python
oecd_bli.info()
```

```python
gdp_per_capita = pd.read_csv("gdp_per_capita_2015.csv", thousands=',', delimiter='\t', encoding='latin1', na_values="n/a")
```

```python
gdp_per_capita.head()
```

```python
gdp_per_capita.info()
```

```python
gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
gdp_per_capita.set_index("Country", inplace=True)
```

```python
gdp_per_capita.head()
```

```python
full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita, left_index=True, right_index=True)
```

```python
full_country_stats.head()
```

```python
full_country_stats.info()
```

```python
full_country_stats.sort_values(by="GDP per capita", inplace=True)
```

```python
country_stats = full_country_stats[["GDP per capita", 'Life satisfaction']]
```

```python
country_stats
```

```python
country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(8,3))
plt.axis([0, 110000, 0, 10])
```

```python
remove_indices = [0, 1, 6, 8, 33, 34, 35]
keep_indices = list(set(range(36)) - set(remove_indices))

sample_data = full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
missing_data = full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[remove_indices]
```

```python
missing_data
```

```python
pos_text_y = (8.5, 4, 2, 9.2, 5, 4, 3)
```

```python
sample_data.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(8,3))
plt.axis([0, 110000, 0, 10])

for i, country in enumerate(missing_data.index):
    pos_data_x, pos_data_y = missing_data.loc[country]
    plt.annotate(country, xy=(pos_data_x, pos_data_y), xytext=(pos_data_x, pos_text_y[i]), ha="center",
           arrowprops=dict(facecolor='black', width=0.5, shrink=0.1, headwidth=5))
    plt.plot(pos_data_x, pos_data_y, "rs")
```

```python
sample_data
```

```python
random_countries = ['Russia', 'Poland', 'Korea', 'New Zealand', 'United Kingdom', 'Australia', 'United States']
```

```python
pos_text_y = (8, 3, 2, 9, 4, 2, 4)
```

```python
sample_data.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(5,3))
plt.axis([0, 61000, 0, 10])
for i, country in enumerate(random_countries):
    pos_data_x, pos_data_y = sample_data.loc[country]
    if country == "United Kingdom": country = "U.K."
    if country == "United States": country = "U.S."
    plt.annotate(country, xy=(pos_data_x, pos_data_y), xytext=(pos_data_x, pos_text_y[i]),
            arrowprops=dict(facecolor='black', width=0.5, shrink=0.1, headwidth=5))
    plt.plot(pos_data_x, pos_data_y, "ro")
plt.xlabel("GDP per capita (USD)")
fig = plt.gcf()
fig.savefig('money_happy_scatterplot.pdf', bbox_inches='tight')
```

```python
import numpy as np

sample_data.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(5,3))
plt.xlabel("GDP per capita (USD)")
plt.axis([0, 60000, 0, 10])
X = np.linspace(0, 60000, 1000)
plt.plot(X, 2*X/100000, "r")
plt.text(40000, 2.7, r"$\theta_0 = 0$", fontsize=14, color="r")
plt.text(40000, 1.8, r"$\theta_1 = 2 \times 10^{-5}$", fontsize=14, color="r")
plt.plot(X, 8 - 5*X/100000, "g")
plt.text(5000, 9.1, r"$\theta_0 = 8$", fontsize=14, color="g")
plt.text(5000, 8.2, r"$\theta_1 = -5 \times 10^{-5}$", fontsize=14, color="g")
plt.plot(X, 4 + 5*X/100000, "b")
plt.text(5000, 3.5, r"$\theta_0 = 4$", fontsize=14, color="b")
plt.text(5000, 2.6, r"$\theta_1 = 5 \times 10^{-5}$", fontsize=14, color="b")
fig = plt.gcf()
fig.savefig('money_happy_scatterplot.pdf', bbox_inches='tight')
```

```python
X_data = np.c_[sample_data["GDP per capita"]]
y_data = np.c_[sample_data["Life satisfaction"]]
```

```python
import sklearn.linear_model

# Select a linear model
linear_regression_model = sklearn.linear_model.LinearRegression()

# Train the model
linear_regression_model.fit(X_data, y_data)

# Model parameters
t0, t1 = linear_regression_model.intercept_[0], linear_regression_model.coef_[0][0]
t0, t1
```

```python
sample_data.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(5,3))
plt.xlabel("GDP per capita (USD)")
plt.axis([0, 60000, 0, 10])
X = np.linspace(0, 60000, 1000)
plt.plot(X, t0 + t1*X, "b")
plt.text(5000, 3.1, r"$\theta_0 = 4.85$", fontsize=14, color="b")
plt.text(5000, 2.2, r"$\theta_1 = 4.91 \times 10^{-5}$", fontsize=14, color="b")
fig = plt.gcf()
fig.savefig('best_fit_model_plot.pdf', bbox_inches='tight')
```

```python
gdp_per_capita.sort_values(by="GDP per capita")["GDP per capita"][130:170]
```

```python
query_gdp_per_capita = gdp_per_capita.loc["Malta"]["GDP per capita"]
print(query_gdp_per_capita)

query_predicted_life_satisfaction = linear_regression_model.predict([[query_gdp_per_capita]])[0][0]
query_predicted_life_satisfaction
```

```python
sample_data.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(5,3), s=1)
plt.xlabel("GDP per capita (USD)")
X = np.linspace(0, 60000, 1000)
plt.plot(X, t0 + t1*X, "b")
plt.axis([0, 60000, 0, 10])
plt.text(5000, 7.5, r"$\theta_0 = 4.85$", fontsize=14, color="b")
plt.text(5000, 6.6, r"$\theta_1 = 4.91 \times 10^{-5}$", fontsize=14, color="b")
plt.plot([query_gdp_per_capita, query_gdp_per_capita], [0, query_predicted_life_satisfaction], "r--")
plt.text(25000, 5.0, r"Prediction = 5.97", fontsize=14, color="b")
plt.plot(query_gdp_per_capita, query_predicted_life_satisfaction, "ro")
fig = plt.gcf()
fig.savefig('prediction_plot.pdf', bbox_inches='tight')
```

```python
import sklearn.neighbors

# Select a k-NN regression model (k=3)
knn_regression_model = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)

# Train the model
knn_regression_model.fit(X_data, y_data)

# Make a prediction for a country
X_new = [[query_gdp_per_capita]]  # GDP per capita of that country
print(knn_regression_model.predict(X_new))
```
