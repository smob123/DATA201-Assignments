The dataset used in this lab is file ``citibike.csv`` containing information about bicycle rental at a particular station. The task we want to solve is to predict for a given time and day how many people will rent a bike at that station.

Let's get an overview of the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_mine = pd.read_csv("citibike.csv")
```

```python
data_mine.head(5)
```

```python
data_mine["start station name"].unique()
```

As you can see, all the records about bicycle rental at only one station (E 2 St & Avenue B).

We will create a new column `one`, resample the data by `starttime` in `3hour` frequency, compute the number of bicycle rentals in each bin, and store that value into column `one`.

```python
data_mine['one'] = 1
data_mine['starttime'] = pd.to_datetime(data_mine.starttime)
```

```python
data_mine.head(5)
```

```python
data_starttime = data_mine.set_index("starttime")
```

```python
data_starttime.head(5)
```

```python
data_resampled = data_starttime.resample("3h").sum().fillna(0)
```

```python
data_resampled.head(5)
```

```python
citibike = data_resampled.one
```

```python
citibike.head(5)
```

Now `citibike` is the data for our task.

**Visualization of the rental frequencies for the whole month:**

```python
plt.figure(figsize=(10, 3))
xticks = pd.date_range(start=citibike.index.min(), end=citibike.index.max(), freq='D')
plt.xticks(xticks, xticks.strftime("%a %m-%d"), rotation=90, ha="left")
plt.plot(citibike, linewidth=1)
plt.xlabel("Date")
plt.ylabel("Rentals");
```

The only feature that we are using in our prediction task is the date and time when a particular number of rentals occurred. So, the input feature is the date and time, e.g., `2015-08-01 00:00:00`, and the output is the number of rentals in the following three hours (`3` in this case).

A common way that dates are stored on computers is using POSIX time, which is the number of seconds since `January 1970 00:00:00` (the beginning of Unix time). We will use this single integer feature as our data representation:

```python
# convert to POSIX time by dividing by 10**9
X = citibike.index.astype("int64").values.reshape(-1, 1) // 10**9

# extract the target values (number of rentals)
y = citibike.values
```

```python
X.shape
```

```python
print(X[:5], y[:5])
```

```python
(X[1]-X[0]) / 3600
```

Looking at the data in the plot above, we can clearly distinguish day and night for each 24-hour interval. The patterns for weekdays and weekends also seem to be quite different. When evaluating a prediction task on a **time series** like this, we usually want to **learn from the past** and **predict for the future**. This means when doing a split into a training and a test set, we want to use all the data up to a certain date as the training set and all the data past that date as the test set.

We will use the first 184 data points, corresponding to the first 23 days, as our training set, and the remaining 64 data points, corresponding to the remaining 8 days, as our test set. *Note that it is **not appropriate** to use **random** spliting on time-series data.*

Below is a function to evaluate and plot a regressor on a given feature set using the first `n_train` (=184 by default) data points for training, the rest for testing.

```python
def eval_on_features(features, target, regressor, n_train=184):
    # split the given features into a training and a test set
    X_train, X_test = features[:n_train], features[n_train:]
    # also split the target array
    y_train, y_test = target[:n_train], target[n_train:]
    
    regressor.fit(X_train, y_train)
    print("Test-set R^2: {:.2f}".format(regressor.score(X_test, y_test)))
    y_pred = regressor.predict(X_test)
    y_pred_train = regressor.predict(X_train)
    
    plt.figure(figsize=(10, 3))
    plt.xticks(range(0, len(X), 8), xticks.strftime("%a %m-%d"), rotation=90, ha="left")

    plt.plot(range(n_train), y_train, label="train")
    plt.plot(range(n_train, len(y_test) + n_train), y_test, '-', label="test")
    plt.plot(range(n_train), y_pred_train, '--', label="prediction train")

    plt.plot(range(n_train, len(y_test) + n_train), y_pred, '--', label="prediction test")
    plt.legend(loc=(1.01, 0))
    plt.xlabel("Date")
    plt.ylabel("Rentals")
```

Let's train a random forest regressor and test it:

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
eval_on_features(X, y, regressor)
```

The predictions on the training set are quite good; however, for the test set, *a constant line is predicted*. The $R^2$ is 0.04, which means that the model learned nothing.

This is due to the combination of our feature and the random forest. The values of the POSIX time feature for the test set are outside of the range of the feature values in the training set. Random forests, and other tree-based regressors, **cannot extrapolate** to feature ranges outside the training set. The result is that *the model simply predicts the target value of the closest point in the training set*, which is the last time it observed any data. Note that this problem has already been discussed in our previous lecture about Decision Trees.

Let's fix this issue using our *expert knowledge*. From looking at the rental figures in the training data, two factors seem to be very important: *the time of day* and *the day of the week*. So, let's add these two features.

We can't really learn anything from the POSIX time, so we drop that feature. First, let's use only the hour of the day.

```python
X_hour = citibike.index.hour.values.reshape(-1, 1)
```

**Write code to made prediction using a random forest regressor with only the hour of the day:**

```python
eval_on_features(X_hour, y, regressor)
```

The $R^2$ is already much better, but the predictions still miss the weekly pattern.

**Write code to made prediction using a random forest regressor with day of week and hour of day features:**

Hint: use `citibike.index.dayofweek.values` to get the *day of week* feature.

```python
X_hour_week = np.hstack([citibike.index.dayofweek.values.reshape(-1, 1),
                         citibike.index.hour.values.reshape(-1, 1)])
eval_on_features(X_hour_week, y, regressor)
```

The predictive performance is quite good now. However, since we are using only 2 features, this does not require a complex model like a random forest. Let's try a simpler model.

**Write code to made prediction using a `LinearRegression` with day of week and hour of day features:**

```python
from sklearn.linear_model import LinearRegression
eval_on_features(X_hour_week, y, LinearRegression())
```

`LinearRegression` works much worse. The reason is that we encoded day of week and time of day using integers, which are
interpreted as categorical variables. Therefore, the linear model can only learn a linear function of the time of day, and it learned that *later in the day, there are more rentals*.

**Write code using `OneHotEncoder` to transform the data with day of week and hour of day features to `X_hour_week_onehot`:**

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
X_hour_week_onehot = enc.fit_transform(X_hour_week).toarray()
```

```python
X_hour_week_onehot[:5]
```

**Write code to evaluate a `Ridge` model using default parameters on `X_hour_week_onehot`:**

```python
from sklearn.linear_model import Ridge
eval_on_features(X_hour_week_onehot, y, Ridge())
```

As a linear model, now the model learns one coefficient for each *day of the week*, and one coefficient for each *time of the day*. We can allow the model to learn one coefficient for each combination of day and time of day:

```python
from sklearn.preprocessing import PolynomialFeatures
poly_transformer = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_hour_week_onehot_poly = poly_transformer.fit_transform(X_hour_week_onehot)
```

**Write code to evaluate the Ridge model on `X_hour_week_onehot_poly`**:

```python
lr = Ridge()
eval_on_features(X_hour_week_onehot_poly, y, lr)
```

This transformation finally yields a model that performs similarly well to the random forest. A big benefit of this model is that it is very clear what is learned: one coefficient for each day and time. We can simply plot the coefficients learned by the model, something that would not be possible for the random forest.

First, we create feature names for the hour and day features:

```python
hour = ["%02d:00" % i for i in range(0, 24, 3)]
day = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
features = day + hour
```

Then we name all the interaction features extracted by `PolynomialFeatures`, using the `get_feature_names` method, and keep only the features with *nonzero coefficients*:

```python
features_poly = poly_transformer.get_feature_names(features)
features_nonzero = np.array(features_poly)[lr.coef_ != 0]
coef_nonzero = lr.coef_[lr.coef_ != 0]
```

Now we can visualize the coefficients learned by the linear model:

```python
plt.figure(figsize=(15, 3))
plt.plot(coef_nonzero, 'o')
plt.xticks(np.arange(len(coef_nonzero)), features_nonzero, rotation=90)
plt.xlabel("Feature name")
plt.ylabel("Feature magnitude");
```
